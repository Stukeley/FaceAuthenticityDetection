```python
import numpy as np

def ReLU(x):
    return np.maximum(0, x)


def Softmax(x):
    return np.exp(x) / np.exp(x).sum(axis=0)
```

```python
# Wykres funkcji ReLU
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

plt.plot(x, ReLU(x))
plt.title('ReLU')
plt.show()
```

```python
# Wykres funkcji Softmax
x = np.linspace(-10, 10, 100)

y = Softmax(x)

plt.plot(x, Softmax(x))
plt.title('Softmax')
plt.show()

print(np.sum(y))
```
