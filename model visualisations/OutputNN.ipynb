```python
# TensorFlow
def ProbabilityNN(input_shape=(4,)):  # The input shape has 5 features
    # Input layer
    inputs = tf.keras.Input(shape=input_shape)

    # Fully connected layer 1 with ReLU activation
    x = layers.Dense(10, activation='relu')(inputs)

    # Fully connected layer 2 (output layer)
    x = layers.Dense(1)(x)

    # Sigmoid activation for probability output
    outputs = layers.Activation('sigmoid')(x)

    # Create the model
    model = tf.keras.Model(inputs=inputs, outputs=outputs)

    return model
```

```python
# Build the model using the functional API
model = ProbabilityNN()

# Print the model summary
model.summary()

# Visualize the model architecture
plot_model(model, to_file='probability_nn_functional.png', show_shapes=True)
```
